```python
url = "http://127.0.0.1:5000/predict"
```

```python
data = {
    "Pregnancies": 6,
    "Glucose": 148,
    "BloodPressure": 72,
    "SkinThickness": 35,
    "Insulin": 0,
    "BMI": 33.6,
    "DiabetesPedigreeFunction": 0.627,
    "Age":50
}
```

```python
response = requests.post(url, json=data)
print(response)
print(response.json())
```

```python
from flask import Flask, request, jsonify
import threading

app = Flask(__name__)

@app.route('/predict', methods=['POST'])
def predict():
    data = request.json
    # Simulate a prediction based on Glucose level
    if data and 'Glucose' in data and data['Glucose'] > 140:
        prediction_result = {"prediction": "Diabetes"}
    else:
        prediction_result = {"prediction": "No Diabetes"}
    return jsonify(prediction_result)

def run_flask_app():
    app.run(host='0.0.0.0', port=5000, debug=False, use_reloader=False)
if __name__ == '__main__':
    print("Starting Flask server on port 5000...")
    thread = threading.Thread(target=run_flask_app)
    thread.daemon = True
    thread.start()
    print("Flask server started in a background thread.")
```
